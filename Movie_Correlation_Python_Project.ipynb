```python
#compare the budget and gross revenue
plt.scatter(x=df['budget'],y=df['gross'],color='black')
plt.title('Budget vs Gross')
plt.xlabel('budget')
plt.ylabel('gross')
plt.show()
```

```python
#correlation between budget vs gross
sns.regplot(x='budget',y='gross',data=df, scatter_kws={'color':'#24D4CD'},line_kws={'color':'black'})

##"scatter_kws" to provide style info for the points.[s for size], "line_kws" to provide style info for the line
```

```python
#lets start looking at correlation
#df.corr():finds the pairwise correlation of each column in a DataFrame
#different methods of correlation:pearson(default),kendall,spearman
df.corr()
#movies.corr(method='kendall')
```

### Budget has high correlation with gross- assumption is right

```python
corr_matrix = df.corr()
sns.heatmap(corr_matrix,annot=True)
plt.title("Correlation between numeric features")
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#look at company
df.head()
```

```python
#company
#company is not numeric lets convert it to numeric representation
df['company'] = df['company'].astype('category')
```

```python
df_category = df
```

```python
#cat codes are for categorical data,it returns a series of codes for each category
df_category['company'] = df_category['company'].cat.codes
```

```python
df_category.corr()
```

```python
corr_matrix = df_category.corr()
sns.heatmap(corr_matrix,annot=True)
plt.title("Correlation between numeric features")
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#organize to see which have the highest correlation by unstacking
df_category.corr()
```

```python
corr_matrix=df_category.corr()
```

```python
corr_pairs=corr_matrix.unstack()
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_correlation = sorted_pairs[(sorted_pairs)>0.5]
high_correlation
```

### votes and budget have the highest correlation with gross
### company has low correlation -> assumption was wrong
